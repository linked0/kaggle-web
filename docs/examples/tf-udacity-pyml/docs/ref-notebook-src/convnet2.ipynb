```python
import tensorflow as tf
features = tf.range(-2, 3)
```

```python
sess = tf.InteractiveSession()
```

```python
sess.run([features, tf.nn.relu(features)])
```

```python
features = tf.to_float(tf.range(-1,3))
sess.run([features, tf.sigmoid(features)])
```

```python
features = tf.to_float(tf.range(-1,100, 10))
sess.run([features, tf.tanh(features)])
```

```python
features = tf.constant([-0.1, -0.2, 0.1, 0.2])
```

```python
sess.run([features, tf.nn.dropout(features, keep_prob=0.5)])
```

```python
batch_size = 1
input_height = 3
input_width = 3
input_channels = 1
layer_input = tf.constant([
        [
            [[1.0], [0.2], [1.5]],
            [[0.1], [1.2], [1.4]],
            [[1.1], [0.4], [0.4]]
        ]
    ])
```

```python
kernel = [batch_size, input_height, input_width, input_channels]
max_pool = tf.nn.max_pool(layer_input, kernel, [1,1,1,1], "VALID")
sess.run(max_pool)
```

```python
from scipy.stats.kde import gaussian_kde
from scipy.stats import norm
from numpy import linspace, hstack
from pylab import plot, show, hist

sample1 = norm.rvs(loc=-1.0, scale=1, size=320)
sample2 = norm.rvs(loc=2.0, scale=0.6, size=320)
sample = hstack([sample1, sample2])
probDensityFun = gaussian_kde(sample)
plt.title('KDE Demonstration using Scipy and Numpy', fontsize=20)
x = linspace(-5, 5, 200)
plot(x, probDensityFun(x), 'r')
```

```python
hist(sample, normed=1, alpha=0.45, color='purple')
show()
```
